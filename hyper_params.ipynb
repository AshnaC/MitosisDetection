```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
thresholds =[0.3, 0.4, 0.5]
epochs = [200, 300, 400, 500]
batch_size = [8,12]
patches_per_slide = [10, 15, 20]

patch_size = [256, 512]
```

```python
def show_bar_plot(title, xlabel, labels, metrics):
    f1 = metrics['f1']
    precision = metrics['precision']
    recall = metrics['recall']
    x = np.arange(len(f1))
    width = 0.25
    fig = plt.figure(figsize=(10, 6))

    plt.title(title)
    

    plt.bar(x, np.array(f1), width, label='F1 Score', color="plum")
    plt.bar(x + width, np.array(precision), width, label='Precision', color="khaki")
    plt.bar(x + 2* width, np.array(recall), width, label='Recall', color="lightskyblue")


    for i in range(len(x)):
        height_out = f1[i] + 0.01
        plt.text(i, height_out, f1[i], ha='center')

        height_inp = precision[i]  + 0.01
        plt.text( i+ width, height_inp, precision[i] , ha='center')
        
        height_inp = recall[i]  + 0.01
        plt.text( i+ 2 *width, height_inp, recall[i] , ha='center')
    
    # plt.legend(loc='lower left', fancybox=True, shadow=True, )
    tick_position = x + width
    plt.xticks(tick_position, labels)
    fig.set_facecolor('#FFF')
    plt.xlabel(xlabel)
    plt.legend(loc='lower left', fancybox=True, shadow=True, )
    plt.show()
```

```python
metrics = {
    'f1':[0.606, 0.549, 0.423],
    'precision':[0.656, 0.745, 0.846],
    'recall':[0.563, 0.435, 0.282]
}
show_bar_plot('Test set evaluation on accept score threshold', 'Score thresholds', thresholds, metrics)
```

```python
metrics = {
    'f1':[0.549,0.606, 0.602, 0.622],
    'precision':[0.523,0.656, 0.585, 0.701],
    'recall':[0.578,0.563, 0.620, 0.559]
}
show_bar_plot('Test set evaluation on number of epochs', 'Number of training epochs', epochs, metrics)
```
